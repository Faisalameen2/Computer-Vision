```python
import cv2
import numpy as np
```

```python
#HSV, H: Hue, S: Saturation, V: Value. Each one has lower and upper values

while (True):
    frame = cv2.imread("Data/Smarties-UK-Candies.jpg")
    hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
    
    l_b = np.array([110,50,50]) #l_b is lower blue values, 110,50,50 is HSV fot blue
    u_b = np.array([130,255,255]) #u_b is upper blue values, 130,255,255 is HSV for blue
    
    mask = cv2.inRange(hsv, l_b, u_b) #Threshold the HSV image to get the blue
    #MASK means, the blue will be white, but other colors will be threshold to black
    
    result = cv2.bitwise_and(frame, frame, mask = mask)
    
    frame = cv2.resize(frame, (250,250))
    mask = cv2.resize(mask, (250, 250))
    result = cv2.resize(result, (250,250))
    cv2.imshow("Image", frame)
    cv2.imshow("mask", mask)
    cv2.imshow("result", result)
    
    k = cv2.waitKey(0)
    if k == 27:
        break
        
    cv2.destroyAllWindows()
```
