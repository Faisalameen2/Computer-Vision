```python
import cv2
import numpy as np

img = cv2.imread("Data/messi.jpg")
grey_img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
template = cv2.imread("Data/Messi Face.jpg", 0)

w,h = template.shape[::-1] #-1: go back to video to remeber the idea

res = cv2.matchTemplate(grey_img, template, cv2.TM_CCORR_NORMED)
cv2.imshow("result", res)
print(res)

threshold = .99
loc = np.where(res>=threshold)
print(loc)

for pt in zip(loc[1], loc[0]):
    cv2.rectangle(img, pt, (pt[0] + w, pt[1] + h), (0,0,255), 2)

cv2.imshow("img", img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
