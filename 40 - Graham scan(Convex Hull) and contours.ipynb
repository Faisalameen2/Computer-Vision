The idea behind convex hull is to draw a polygon around the points, which must cover all the points

convex corners always less than 180 degree

# 

the Concave always larger than 180 degree. the point in the middle is concave

# 

This is the difference between convex and concave 

```python
import cv2
#import matplotlib.pyplot as plt

image = cv2.imread("Data/cars.png")
image = cv2.resize(image,(650,500))

img = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

ret, thresh = cv2.threshold(img, 50, 255, 0)

contours, hierarchy = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)

for i in range(len(contours)):
    hull = cv2.convexHull(contours[i])
    cv2.drawContours(image, [hull], -1, (0,255,0),2)
    
cv2.imshow("ConvexHull", image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```
